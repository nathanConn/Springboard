Prints the 10 major project themes

```python
theme_names = pd.merge(json_normalize(j, 'mjtheme_namecode').groupby('code').count().sort_values('name', ascending=False)[:10].reset_index(),json_normalize(j, 'mjtheme_namecode'),
         how='left',on='code')

theme_names.name_y.value_counts().nlargest(10)
```

Create a dataframe with the missing names filled in

```python
theme_names['name_y'].replace('', np.nan, inplace=True)
tn1 = theme_names.copy()

tn1.dropna(subset=['name_y'], inplace=True)

theme_names_sorted = theme_names.sort_values(['name_x', 'name_y'])
theme_names_sorted.fillna(method="ffill",inplace=True)
theme_names_sorted
```
